# Rename files

## Edible

```python
# function to rename multiple files
# amanita_muscaria
def main():
   
    folder = "data/images/POC Mushrooms/edible/Morchella_esculenta"
    for count, filename in enumerate(os.listdir(folder)):
        dst = f"amanita_muscaria {str(count)}.jpg"
        src =f"{folder}/{filename}"  # foldername/filename, if .py file is outside folder
        dst =f"{folder}/{dst}"
         
        # rename() function will
        # rename all the files
        os.rename(src, dst)
 
# Driver Code
if __name__ == '__main__':
     
    # calling main() function
    main()
```

# Create new dataframe



```python
import pandas as pd
```

```python
images = os.listdir('data/images/POC Mushrooms/edible/Amanita_citrina')
```

```python
images[:5]
```

```python
images_df = pd.DataFrame({'species':[i.split('.')[0] for i in images],
                         'path':['data/images/POC Mushrooms/edible/' + i for i in images]})
                                 
# remove any digits from string
images_df['species'] = images_df['species'].str.replace('\d+', '')

# remove any spaces from strings
images_df['species'] = images_df['species'].str.strip()
```

```python
images_df
```

```python
# reading in the tabular data containing mushroom features
df_tabular = pd.read_csv('data/tabular/tabular_data')
```

```python
df_tabular
```

```python
df_tabular.loc[df_tabular["species"] == "fly agaric", "species"] = 'amanita_muscaria'
```

```python
df_tabular
```

```python
merged_df = pd.merge(left = images_df, right = df_tabular, on='species', how='inner')
```

```python
merged_df
```
